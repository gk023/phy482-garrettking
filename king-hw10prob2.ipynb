```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
k0=13
c=3e17
s=0.013
x=np.arange(-300.0,300.0,1.0)
t=0

f = np.exp(-(x-c*t)**2*s**2/2)*np.cos(k0*(x-c*t))

plt.figure(figsize=(16,8))
plt.plot(x,f,color="green")
plt.xlabel("x (nm)")
plt.ylabel("(f(x))")
plt.show()
```

```python
#I converted c to nm and assumed if it was a t*10^-15 s pulse that delta x would be c*t=300nm. Using the relationship between
#sigma and delta x found before, I calculated that parameter. From there, I plotted the function.
```
