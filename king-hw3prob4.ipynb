```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
a=0.01
v=2.5
w=0.05
B0 = 0.2

t=np.arange(0.0,0.005,1*10e-6)
x=v*t

E=np.zeros(len(t))

```

```python
for i in range(len(t)):
    E[i]=-w*((B0+a*t[i])*v+a*x[i])
    
fig=plt.figure(figsize=(8,8))

plt.plot(t,E)
plt.show()
```

```python
a1=0.0001
E1=np.zeros(len(t))


for i in range(len(t)):
    E1[i]=-w*((B0+a1*t[i])*v+a1*x[i])
    
fig=plt.figure(figsize=(8,8))

plt.plot(t,E1)
plt.show()
```

```python
a2=0.1
E2=np.zeros(len(t))

for i in range(len(t)):
    E2[i]=-w*((B0+a2*t[i])*v+a2*x[i])
    
fig=plt.figure(figsize=(8,8))

plt.plot(t,E2)
plt.show()
```

```python
v1=0.0025
E3=np.zeros(len(t))

x1=v1*t

for i in range(len(t)):
    E3[i]=-w*((B0+a*t[i])*v1+a*x1[i])
    
fig=plt.figure(figsize=(8,8))

plt.plot(t,E3)
plt.show()
```

```python
v2=10
E4=np.zeros(len(t))

x2=v2*t

for i in range(len(t)):
    E4[i]=-w*((B0+a*t[i])*v2+a*x2[i])
    
fig=plt.figure(figsize=(8,8))

plt.plot(t,E4)
plt.show()
```

```python
v3=0.0000025
E5=np.zeros(len(t))

x3=v3*t


for i in range(len(t)):
    E5[i]=-w*((B0+a*t[i])*v3+a*x3[i])
    
fig=plt.figure(figsize=(8,8))

plt.plot(t,E5)
plt.show()
```

```python
#Either way, the emf seems to have the same qualitative behavior as the bar moves into the field, but the size
#of the parameters determines the magnitude of the emf at time zero and the magnitude of the changes in the emf over
#a given time interval. In the case where a is small compared to v, the magnitude of the changes in the emf are smaller. If
#v is large compared to a, then the size of the jumps in the magnetic field are larger. Either way, I think that it is the
#combination of the two that makes the behavior linear. If either a or v goes to, the behavior is pretty much
#like a constant value.
```
