```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
x = np.arange(-5,5.5,0.5)
y = np.arange(-5,5.5,0.5)
X,Y = np.meshgrid(x,y)
```

```python
k = 9e9
q = 1.6e-19

Elim=1e-8

Epx = k*q*X/(X**2+Y**2)**(3/2)
Epy = k*q*Y/(X**2+Y**2)**(3/2)

mEpx = np.ma.masked_array(Epx, np.abs(Epx) > Elim)
mEpy = np.ma.masked_array(Epy, np.abs(Epy) > Elim)

fig=plt.figure(figsize=(8,8))
pt=plt.quiver(X[::3],Y[::3],mEpx[::3],mEpy[::3], units = 'inches' )

plt.show()
```

```python
Ed1x = k*q*X/(X**2+(Y-0.5)**2)**(3/2)
Ed1y = k*q*(Y-0.5)/(X**2+(Y-0.5)**2)**(3/2)

Ed2x = -k*q*X/(X**2+(Y+0.5)**2)**(3/2)
Ed2y = -k*q*(Y+0.5)/(X**2+(Y+0.5)**2)**(3/2)

Edx = Ed1x + Ed2x
Edy = Ed1y + Ed2y

mEdx = np.ma.masked_array(Edx, np.abs(Edx) > Elim*0.05)
mEdy = np.ma.masked_array(Edy, np.abs(Edy) > Elim*0.05)

fig=plt.figure(figsize=(8,8))
dip=plt.quiver(X[::3],Y[::3],mEdx[::3],mEdy[::3], units = 'inches' )

plt.show()
```

```python
Eq1x = k*q*(X-0.5)/((X-0.5)**2+(Y-0.5)**2)**(3/2)
Eq1y = k*q*(Y-0.5)/((X-0.5)**2+(Y-0.5)**2)**(3/2)

Eq2x = -k*q*(X+0.5)/((X+0.5)**2+(Y-0.5)**2)**(3/2)
Eq2y = -k*q*(Y-0.5)/((X+0.5)**2+(Y-0.5)**2)**(3/2)

Eq3x = k*q*(X+0.5)/((X+0.5)**2+(Y+0.5)**2)**(3/2)
Eq3y = k*q*(Y+0.5)/((X+0.5)**2+(Y+0.5)**2)**(3/2)

Eq4x = -k*q*(X-0.5)/((X-0.5)**2+(Y+0.5)**2)**(3/2)
Eq4y = -k*q*(Y+0.5)/((X-0.5)**2+(Y+0.5)**2)**(3/2)

Eqx = Eq1x + Eq2x + Eq3x + Eq4x
Eqy = Eq1y + Eq2y + Eq3y + Eq4y

mEqx = np.ma.masked_array(Eqx, np.abs(Eqx) > Elim*0.03)
mEqy = np.ma.masked_array(Eqy, np.abs(Eqy) > Elim*0.03)

fig=plt.figure(figsize=(8,8))
quad=plt.quiver(X[::3],Y[::3],mEqx[::3],mEqy[::3], units = 'inches' )

plt.show()
```
