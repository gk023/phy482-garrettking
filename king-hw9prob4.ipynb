```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
x = np.arange(0,91,1)

b=2.0

a=np.sqrt(1-(np.sin(np.pi*x/180)/b)**2)/np.cos(np.pi*x/180)

R=(1-a*b)**2/(1+a*b)**2

T=(4*a*b)/(1+a*b)**2

fig=plt.figure()

plt.plot(x,R,label="R")
plt.plot(x,T,label="T")
plt.plot(x,R+T,label="R+T")
plt.xlabel("Theta_I (deg)")
plt.ylabel("Dimensionless")
plt.legend(loc="best")
plt.show()
```

```python
#This would be part 4. The plot shows that R+T is constantly 1 at every angle from 0 to 90. This makes sense, since R+T
#is the sum of the ratios of the intensities of the reflected and transmitted wave over the incident wave. The fact that
#this is always 1 is a statement of energy conservation. If it weren't one, the intensity, which is related to the energy,
#would be lost from the incident wave to the two resulting waves. Therefore, this should always be 1, because the energy that
#we start with all comes back out in this case.
```

```python
r=(a*b-1)/(1+a*b)
t=2/(1+a*b)

fig=plt.figure()

plt.plot(x,r,label="Er/Ei")
plt.plot(x,t,label="Et/Ei")
plt.xlabel("Theta_I (deg)")
plt.ylabel("Dimensionless")
plt.legend(loc="best")
plt.show()
```

```python
#Based on the plot above, it does not appear that there is a Brewster angle for this case. There is always a reflected wave
#amplitude and it never goes to 0. In both Griffiths' example and this plot, Et/Ei gets smaller with increasing incident
#angle and Er/Ei grows with it. At 90 degrees, there is only a reflected amplitude and notihg is transmitted. The reflected
#wave always has a positive amplitude in the case that I am plotting, where Griffiths' example had portions with a negative
#reflected wave amplitude.
```
