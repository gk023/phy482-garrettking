```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
R = 10
L = 10e-3

x = np.arange(0,5000,25)

trans = R/np.sqrt(R**2+x**2*L**2)

fig=plt.figure(figsize=(16,8))
plt.plot(x,trans)
plt.xlabel("Angular Frequency")
plt.ylabel("Vout/Vin")
plt.show()
```

```python
#This filter would be low pass, since the ratio of the output to input goes from 1 near zero down to zero as the
#frequency goes to infinitiy.
```

```python
w = 250

t=np.arange(0,0.05,5e-4)

Vin = 0
Vout = 0

kmax = 5

k=1

while(k < kmax):
    Vin = Vin + 4/(np.pi)*np.sin((2*k-1)*w*t)/(2*k-1)
    Vout = Vout + R/np.sqrt(R**2+((2*k-1)*w)**2*L**2)*4/(np.pi)*np.sin((2*k-1)*w*t)/(2*k-1)
    k=k+1
    
fig=plt.figure(figsize=(16,8))
plt.plot(t,Vin,label="input")
plt.plot(t,Vout,label="output")
plt.legend(loc="best",fontsize="x-large")
plt.xlabel("time")
plt.ylabel("Voltage")
plt.show()
```

```python
#As the number of sine waves, and thus the frequecies included, get higher, we see that the
#input turns into a square wave while the output stays rounded. This is because the
#higher frequency signals used to smooth out the noise in the square wave are filtered out
#in the output wave. This difference in weighting caused by the filter results in
#the difference in waveforms.
```
