```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
V = 5
R = 250
C = 50e-9
L = 25e-3

step=100
f = np.arange(100,100000+step,step)

I = V*np.sqrt(1/R**2+(2*np.pi*f*C-1/(2*np.pi*f*L))**2)

fig=plt.figure(figsize=(8,8))
plt.plot(f,I)
plt.xlabel("Frequency (Hz)")
plt.ylabel("Magnitude of Current (A)")
plt.xlim([100,100000])
plt.xscale("log")
plt.show()
```

```python
#The magnitude of the current has a minimus value at 4500 Hz, which is 1/sqrt(LC)
```
